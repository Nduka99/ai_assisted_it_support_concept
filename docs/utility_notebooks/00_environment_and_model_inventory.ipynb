# IT-SUPPORT: Environment and Model Inventory

Run this notebook first. It checks the project paths, local model registry, and GPU status without loading large models.

```python
from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().resolve()
if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent

SRC = PROJECT_ROOT / "src"
if str(SRC) not in sys.path:
    sys.path.insert(0, str(SRC))

PROJECT_ROOT
```

```python
from it_support.config import LOCAL_MODELS, TARGET_DOWNLOADS

rows = []
for model in LOCAL_MODELS.values():
    path = model.path
    if path.exists() and path.is_file():
        size = f"{path.stat().st_size / (1024 ** 3):.2f} GB"
    elif path.exists() and path.is_dir():
        size = "directory"
    else:
        size = "missing"
    rows.append({
        "key": model.key,
        "role": model.role,
        "backend": model.backend,
        "exists": model.exists,
        "size": size,
        "path": str(path.relative_to(PROJECT_ROOT)),
    })

rows
```

```python
print("Target downloads to consider later:")
for key, reason in TARGET_DOWNLOADS.items():
    print(f"- {key}: {reason}")
```

```python
import subprocess

try:
    result = subprocess.run(["nvidia-smi"], capture_output=True, text=True, check=False)
    print(result.stdout or result.stderr)
except FileNotFoundError:
    print("nvidia-smi not found on PATH")
```
